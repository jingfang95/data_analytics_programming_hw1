```python
from numpy.random import choice

def comp_brain(history): #1 = paper, 2 = sissor, 3 = rock
    
    options = [1,2,3];  
    total = history[0] + history[1] + history[2];
    
    if history == [0, 0, 0]:
        draw = choice(options, 1,p=[1/3,1/3,1/3]);
        
    else:
        p_distribution = [history[0]/total, history[1]/total, history[2]/total]
        draw = choice(options, 1, p = p_distribution);
        
    if draw == 1:
        pick = 2;
    elif draw == 2:
        pick = 3
    elif draw == 3:
        pick = 1;
    
    return (pick)

def human_input(value):
    
    if value == 'Rock':
        choice = 3;
    elif value == 'Paper':
        choice = 1;
    elif value == 'Sissor':
        choice = 2
    elif value == 'Quit':
        choice = 0;
    else: 
        choice = 100
        
    return (choice);

def main():
    storage = [0, 0, 0];
    count1 = 0;
    count2 = 0;
    
    c_choice = comp_brain(storage);    
    p_choice = input("The computer has made its choice. Please enter your choice as 'Rock', 'Paper', 'Sissor' or 'Quit': ");  
    check = human_input (p_choice);   
    
    while check == 100:
        print ("Invalid input. Please try again.")
        p_choice = input("The computer has made its choice. Please enter your choice as 'Rock', 'Paper', 'Sissor' or 'Quit': ");
        check = human_input(p_choice)
        
    if check == 1:
        storage[0] += 1;
    elif check == 2:
        storage[1] += 1;
    elif check == 3:
        storage[2] += 1;
        
    if check != 0:
        count1 += 1;
    
    if c_choice == 1:
        if check == 1:
            print ("Draw!");
        
        if check == 2:
            print ("Computer played paper, you won!")
            count2 += 1;
            
        if check == 3:
            print ("Computer played paper, you lost!")
    
    if c_choice == 2:
        if check == 1:
            print ("Computer played sissor. you lost!");
        
        if check == 2:
            print ("Draw!")
            
        if check == 3:
            print ("Computer played sissor. you won!");
            count2 += 1;
             
    if c_choice == 3:
        
        if check == 1:
            print ("Computer played rock, you won!");
            count2 += 1
            
        if check == 2:
            print ("Computer played rock, you lost!")
            
        if check == 3:
            print ("Draw!")
   
    while check != 0:
        c_choice = comp_brain(storage);
        p_choice = input("The computer has made its choice. Please enter your choice as 'Rock', 'Paper', 'Sissor' or 'Quit': ");
        check = human_input (p_choice); 
        
        while check == 100:
            print ("Invalid input. Please try again.")
            p_choice = input("The computer has made its choice. Please enter your choice as 'Rock', 'Paper', 'Sissor' or 'Quit': ");
            check = human_input (p_choice);
            
        if check == 1:
            storage[0] += 1;
        elif check == 2:
            storage[1] += 1;
        elif check == 3:
            storage[2] += 1;
            
        if check != 0:
            count1 += 1

        if c_choice == 1:
            if check == 1:
                print ("Draw!");
        
            if check == 2:
                print ("Computer played paper, you won!")
                count2 += 1;
            
            if check == 3:
                print ("Computer played paper, you lost!")
    
        if c_choice == 2:
            if check == 1:
                print ("Computer played sissor. you lost!");
        
            if check == 2:
                print ("Draw!")
            
            if check == 3:
                print ("Computer played sissor. you won!");
                count2 += 1;
             
        if c_choice == 3:      
            if check == 1:
                print ("Computer played rock, you won!");
                count2 += 1
            
            if check == 2:
                print ("Computer played rock, you lost!")
            
            if check == 3:
                print ("Draw!")
        
    print (str(count1) + " Games played, the player has won " + str(count2) + " game(s).")

main()
```
