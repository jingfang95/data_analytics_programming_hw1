```python
import re
import string
import math
from collections import Counter 

def most_frequent(List):  #Input list, returns the most frequent word's count.
    counter = 0
    num = List[0] 
      
    for i in List: 
        curr_frequency = List.count(i) 
        if(curr_frequency> counter): 
            counter = curr_frequency 
            num = i 
      
    return List.count(num);
  

def full_quotes(): #Returns the full-quotes in the file
    fp = open('quotes.txt', 'r');
    full = [];
    line = fp.readline().rstrip('\n');
    name = fp.readline().rstrip('\n');
    
    while line != '':
        full.append(line + " - " + name);
        line = fp.readline().rstrip('\n');
        name = fp.readline().rstrip('\n');
        
    fp.close()
        
    return full

def words(quote): #input the full quote, go get a list of lowered words.
    res = re.sub('['+string.punctuation+']', '', quote).split()
    words = []
    for word in res:
        words.append(word.lower());
        
    return words

def pos_list(quote, words): #input full quote, words; to get the dictionary of quote:{word:count}}
    dic1 = {i:words.count(i) for i in words}
    dic2 = {quote : dic1};
    
    return dic2;
    
def rev_list(quote, words): #nput full quote, words; to get the dictionary of word:{quote:count}}
    count = {i:words.count(i) for i in words}
    rev_dic = {}
    for j in words:
        rev_dic.update({j : {quote:count[j]}});
        
    return rev_dic;

def TF_IDF (word, quote): #input full quotes list,  the total word list, word and quote.
    
    x = 0;
    full = full_quotes();
    word_list = [];
    
    while x < len (full):
        word_list.append(words(full[x]));
        x += 1;
    
    
    counter_IDF = 0 #count the number of word appears in all the quotes.
    count = 0;
    
    for i in word_list:
        
        if word in word_list[count]:
            counter_IDF += 1
        
        count += 1;
    
    #counter_IDF gives time that a word appears in the whole file.
    IDF = math.log(len(full)/counter_IDF);
    ###########################################################################
    TF = (words(quote).count(word))/(most_frequent(words(quote)))
    ###########################################################################
    TFIDF = TF*IDF
    return TFIDF

def QS_single (word): #Single word search. input word, give quote : TF-IDF
    
    full = full_quotes();
    
    count = 0;
    count2 = 0;
    word_list = [];
    new_dic = {}
    new_dic2 = {}
    quote = ''
    
    pos_list_dic = [];
    rev_list_dic = [];
    
    
    while count < len (full):
        pos_list_dic.append(pos_list(full[count], words(full[count])));
        rev_list_dic.append(rev_list(full[count], words(full[count])));
        word_list.append(words(full[count]));
        count += 1;
    
    while count2 < len(rev_list_dic):
        
        if word in rev_list_dic[count2]:
            quote = full[count2];
            new_dic = rev_list_dic[count2][word];
            for key in new_dic:
                new_dic[key] = TF_IDF(word, quote);
            
            new_dic2.update(new_dic);
        
        new_dic = {};
        count2 += 1
        
    return new_dic2   

def QS_multi(word):
    input_list = [];
    input_list = word;
    count = 1;

    big_list = [];
    
    for item in input_list:
        big_list.append(QS_single(item));
        
    Cdict = Counter(big_list[0]);
    
    while count < len(big_list):
        Cdict += Counter(big_list[count])
        count += 1;
        
    return Cdict
    
def main():
    print("Testing.")
    
    #TF-IDF
    #print (TF_IDF ('roads', 'Two roads diverged in a wood, and I...I took the one less traveled by, and that has made all the difference. - Robert Frost'))
    
    #Single word search.
    #print (QS_single('two'))
    #print (QS_single('roads'))
    
    #Multi word search
    #print(QS_multi(['two', 'roads', 'diverged']));
   
main()
    
```
